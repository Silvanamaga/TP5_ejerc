Ejercicio 2_ Utilice la librer√≠a os para listar todos los directorios dentro del directorio ‚Äú/dev‚Äù

```python
import os

# Para usar la librer√≠a os uso el comando "import". 
# obtengo la lista de todos los elementos en el directorio "/dev" para lo que
# uso os.listdir y lo guardo en una variable elementos_dev
elementos_dev = os.listdir("/dev")

# Luego filtro solo los directorios (carpetas) de la lista
directorios_dev = [elemento for elemento in elementos_dev if os.path.isdir(os.path.join("/dev", elemento))]
#El filtro permite obtener solo los directorios del directorio "/dev", excluyendo los archivos
# Imprimo la lista de directorios dentro de "/dev"
print("Directorios dentro de /dev:")
for directorio in directorios_dev:
    print(directorio)
```

Ejercicio 3. Hacer un programa en python donde tome dos valores de entrada: sueldo y rango. Con ello, calcule la asignaci√≥n que percibir√° el empleado, considerando la siguiente
tabla:
RANGO   F√ìRMULA
1       sueldo x 0.83
2       sueldo x 1.2
3       sueldo x 5

```python
def asignacion_empl(sueldo, rango):
    if rango == 1:
        asignacion = sueldo * 0.83
    elif rango == 2:
        asignacion = sueldo * 1.2
    elif rango == 3:
        asignacion = sueldo * 5
    else:
        asignacion = 0  # Si el rango no es v√°lido, la asignaci√≥n se establece en 0
        print("Rango no v√°lido. Por favor, seleccione un rango v√°lido.")
    return asignacion

# Funci√≥n para interactuar con persona usuario, as√≠ ingresa datos
def entrada():
    sueldo = float(input("Ingrese el sueldo del empleado: "))
    rango = int(input("Ingrese el rango del empleado (1, 2 o 3): "))
    return sueldo, rango

# Funci√≥n principal del programa
def main():
    sueldo, rango = entrada()
    asignacion = asignacion_empl(sueldo, rango)
    if asignacion:
        print("La asignaci√≥n del empleado es:", asignacion)

if __name__ == "__main__":
    main()
```

    üëÅÔ∏è‚Äçüó®Ô∏è elif: Es una abreviatura de "else if". Se utiliza para evaluar otra condici√≥n si la condici√≥n previa (en el if anterior o en otro elif anterior) no se cumple. En otras palabras, elif se ejecuta solo si las condiciones anteriores no son verdaderas. Puedes tener m√∫ltiples bloques elif despu√©s de un if.

    üëÅÔ∏è‚Äçüó®Ô∏è else: Es una cl√°usula opcional que se utiliza despu√©s de todas las cl√°usulas if y elif en una estructura de control if. Se ejecuta cuando ninguna de las condiciones anteriores es verdadera. Puedes tener solo un bloque else despu√©s de todas las cl√°usulas if y elif.

    La funci√≥n asignacion_empl() toma el sueldo y el rango como argumentos y devuelve la asignaci√≥n correspondiente seg√∫n la tabla. 
    La funci√≥n entrada() se utiliza para obtener el sueldo y el rango del usuario.
    La funci√≥n main() llama a las funciones anteriores, calcula la asignaci√≥n y la imprime si el rango es v√°lido.
    Si el rango proporcionado por el usuario no es 1, 2 o 3, el programa imprime un mensaje indicando que el rango no es v√°lido y establecer√° la asignaci√≥n 0.

6. Crea un programa que simule una calculadora simple con funciones para suma,
resta, multiplicaci√≥n y divisi√≥n

```python
def suma(a, b):
    return a + b

def resta(a, b):
    return a - b

def multiplicacion(a, b):
    return a * b

def division(a, b):
    if b != 0:
        return a / b
    else:
        return "Error: Divisi√≥n por cero"

def mostrar_menu():
    print("Calculadora Simple")
    print("1. Suma")
    print("2. Resta")
    print("3. Multiplicaci√≥n")
    print("4. Divisi√≥n")
    print("5. Salir")
    return input("Seleccione una opci√≥n: ")


while True:
    opcion = mostrar_menu()

    if opcion == '5':
        print("Gracias por usar la calculadora.")
        break

    if opcion not in ['1', '2', '3', '4']:
        print("Opci√≥n no v√°lida. Por favor, seleccione una opci√≥n v√°lida.")
        continue

    num1 = float(input("Ingrese el primer n√∫mero: "))
    num2 = float(input("Ingrese el segundo n√∫mero: "))

    if opcion == '1':
        print("Resultado:", suma(num1, num2))
    elif opcion == '2':
        print("Resultado:", resta(num1, num2))
    elif opcion == '3':
        print("Resultado:", multiplicacion(num1, num2))
    elif opcion == '4':
        print("Resultado:", division(num1, num2))
```
